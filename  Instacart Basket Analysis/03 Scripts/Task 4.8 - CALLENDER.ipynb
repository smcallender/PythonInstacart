Q4: Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
#create "max order" column
df_ords_merged['max_order'] = df_ords_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_merged.head(10)
```

```python
#create "loyalty flag" loyal customer
df_ords_merged.loc[df_ords_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#create "loyalty flag" regular customer
df_ords_merged.loc[(df_ords_merged['max_order'] <= 40) & (df_ords_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#create "loyalty flag" new customer
df_ords_merged.loc[df_ords_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#check value counts
df_ords_merged['loyalty_flag'].value_counts()
```

Q6: The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

A. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

B. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
#create spending flag
df_ords_merged['average_price'] = df_ords_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#averages
df_ords_merged['average_price'] = df_ords_merged['average_price'].round(2)
```

```python
df_ords_merged.head()
```

```python
#create spending flag "low spender"
df_ords_merged.loc[df_ords_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#create spending flag "high spender"
df_ords_merged.loc[df_ords_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_merged.head()
```

```python
#check counts
df_ords_merged['spending_flag'].value_counts()
```

Q7: In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

A: If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

B: If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

C: If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
#create order frequency flag
df_ords_merged['median_prior_orders'] = df_ords_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_merged.head()
```

```python
#create order frequency flag "non-frequent customer"
df_ords_merged.loc[df_ords_merged['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#create order frequency flag "regular customer"
df_ords_merged.loc[(df_ords_merged['median_prior_orders'] > 10) & (df_ords_merged['median_prior_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
#create order frequency flag "frequent customer"
df_ords_merged.loc[df_ords_merged['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
